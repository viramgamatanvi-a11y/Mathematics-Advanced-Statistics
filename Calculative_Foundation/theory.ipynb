# 📘 Calculative Foundation

## 📖 Theory Overview

---

## 🔹 1. Vector Representation

A vector is a quantity that has both **magnitude** and **direction**. In this project, each student's subject scores are represented as a vector.

### Example

```text
Student S001 = [Physics, Chemistry, Biology, Geography, Arts]
```

Vectors help represent data points in multidimensional space.

---

## 📏 2. Vector Norms

A norm measures the length or magnitude of a vector.

### 🔸 Norm-1 (Manhattan Norm)

The sum of absolute values of all vector elements.

```text
||x||₁ = Σ|xᵢ|
```

### 🔸 Norm-2 (Euclidean Norm)

The straight-line distance from the origin to the vector.

```text
||x||₂ = √(Σxᵢ²)
```

Norms are used to measure the size of vectors.

---

## 🎯 3. Dot Product and Angle Between Vectors

The dot product measures similarity between two vectors.

```text
a · b = Σ(aᵢbᵢ)
```

The angle between vectors indicates how closely they point in the same direction.

✅ Small angle → More similar

✅ Large angle → Less similar

---

## 🔄 4. Cross Product

The cross product is defined for 3D vectors and produces a vector perpendicular to both input vectors.

### Applications

* Geometry
* Computer Graphics
* Physics

---

## 📌 5. Vector Projection

Projection measures how much of one vector lies in the direction of another vector.

It helps determine similarity and directional contribution.

---

## 🧮 6. Student × Subject Matrix

A matrix is a rectangular arrangement of numbers.

### In This Project

👨‍🎓 Rows represent students

📚 Columns represent subjects

### Example

```text
150 × 5 Matrix
```

Matrices are used for storing and processing data efficiently.

---

## ➕ 7. Matrix Addition

Matrix addition adds corresponding elements from two matrices.

✅ Both matrices must have the same dimensions.

---

## ✖️ 8. Matrix Multiplication

Matrix multiplication combines information from two matrices.

### Applications

* Machine Learning
* Data Transformations
* Computer Graphics

---

## 🔁 9. Matrix Transpose

The transpose swaps rows and columns.

```text
150 × 5
```

becomes

```text
5 × 150
```

Transpose is used in covariance calculations and matrix operations.

---

## 📊 10. Determinant

The determinant is a scalar value associated with a square matrix.

✅ Determinant ≠ 0 → Matrix is invertible.

Determinants help measure scaling effects of linear transformations.

---

## 🔓 11. Matrix Inverse

The inverse matrix reverses the effect of a matrix transformation.

```text
A × A⁻¹ = I
```

where **I** is the Identity Matrix.

Inverse matrices are used to solve systems of equations.

---

## 📐 12. Line, Plane and Hyperplane

### 📍 Line (1D)

Uses one feature or dimension.

**Example:** Physics Scores

### 📍 Plane (2D)

Uses two dimensions.

**Example:** Physics and Chemistry

### 📍 Hyperplane (5D)

Uses all subject scores.

Each student becomes a point in a five-dimensional space.

---

## 🧭 13. Eigenvalues and Eigenvectors

Eigenvectors represent important directions in the data.

Eigenvalues indicate the importance of those directions.

✅ Large eigenvalues correspond to greater variance in the dataset.

These concepts are fundamental to PCA and dimensionality reduction.

---

## ⚙️ 14. LU Decomposition

LU Decomposition factorizes a matrix into:

```text
A = LU
```

Where:

🔹 L = Lower Triangular Matrix

🔹 U = Upper Triangular Matrix

It simplifies matrix computations and equation solving.

---

## 🧩 15. Singular Value Decomposition (SVD)

SVD decomposes a matrix into three components:

```text
A = UΣVᵀ
```

Where:

🔹 U = Left Singular Vectors

🔹 Σ = Singular Values

🔹 Vᵀ = Right Singular Vectors

### Applications

* Recommendation Systems
* Image Compression
* Dimensionality Reduction

---

## 📉 16. Principal Component Analysis (PCA)

PCA reduces the number of dimensions while preserving maximum variance.

### Benefits

✅ Reduces Complexity

✅ Removes Redundancy

✅ Improves Visualization

✅ Helps Machine Learning Models

In this project, subject scores were reduced from **5 dimensions to 2 dimensions**.

---

## 🎯 17. Linear Discriminant Analysis (LDA)

LDA is a supervised dimensionality reduction technique.

It uses class labels to maximize separation between categories.

### Categories Used

🟢 High Performance

🔴 Low Performance

LDA reduces the dataset while improving class separability.
