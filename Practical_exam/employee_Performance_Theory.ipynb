# 📘 Part A – Theory 

## 📊 1️⃣ Mean, Median, and Mode

### ➕ Mean

**Definition:**
Mean is the average of all values in a dataset.

**Example:**
Employee Salaries = ₹30,000, ₹40,000, ₹50,000

**Mean = (30,000 + 40,000 + 50,000) / 3 = ₹40,000**

### 🎯 Median

**Definition:**
Median is the middle value when data is arranged in ascending order.

**Example:**
₹20,000, ₹30,000, ₹40,000, ₹50,000, ₹60,000

**Median = ₹40,000**

### 🔁 Mode

**Definition:**
Mode is the value that occurs most frequently.

**Example:**
₹30,000, ₹40,000, ₹40,000, ₹50,000

**Mode = ₹40,000**

---

## 📏 2️⃣ Range and Variance

### ↔️ Range

**Definition:**
Range is the difference between the maximum and minimum values.

**Example:**
Salaries = ₹30,000, ₹50,000, ₹80,000

**Range = 80,000 − 30,000 = ₹50,000**

### 📈 Variance

**Definition:**
Variance measures how far the values are spread from the mean.

**Example:**
If employee salaries vary greatly, the variance will be high.

---

## 🔔 3️⃣ Normal Distribution and Poisson Distribution

### 📉 Normal Distribution

**Definition:**
A continuous probability distribution with a bell-shaped curve.

**Example:**
Employee salaries in a company often follow a normal distribution.

### 🔢 Poisson Distribution

**Definition:**
A probability distribution used to count how many times an event occurs in a fixed period.

**Example:**
Number of customer complaints received per day.

---

## ↔️ 4️⃣ Skewness

### 📐 Definition

Skewness measures the asymmetry of a distribution.

### 📌 Example

In a company, most employees earn moderate salaries, but a few managers earn very high salaries.

This creates a **positive skew** in salary distribution.

---

## 🎲 5️⃣ Conditional Probability

### 📖 Definition

Conditional probability is the probability of an event occurring given that another event has already occurred.

### 📌 Example

**P(Promotion | Performance Score > 80)**

This means the probability of promotion when an employee's performance score is greater than 80.

---

## 🔗 6️⃣ Independent vs Mutually Exclusive Events

### 🟢 Independent Events

**Definition:**
Two events are independent if the occurrence of one does not affect the other.

**Example:**
An employee's Age and Department.

### 🔴 Mutually Exclusive Events

**Definition:**
Two events are mutually exclusive if they cannot happen at the same time.

**Example:**
An employee cannot be both **Promoted = Yes** and **Promoted = No** at the same time.

---

## 🧠 7️⃣ Bayes Theorem

### 📖 Definition

Bayes Theorem is used to update the probability of an event based on new information.

### 📌 Example

If an employee has a high performance score, Bayes Theorem can help estimate the updated probability of promotion.

---

## 📉 8️⃣ PCA (Principal Component Analysis)

### 📖 Definition

PCA is a dimensionality reduction technique that converts many variables into a smaller number of important components while keeping most of the information.

### 📌 Example

Employee data containing:

* 👤 Age
* 💰 Salary
* ⏰ Working Hours
* 📊 Performance Score

can be reduced to **2 principal components** for easier analysis and visualization.
